```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://patentscope.wipo.int/search/en/search.jsf"
data = {
    "simpleSearchSearchForm": "simpleSearchSearchForm",
    "docId": "WO2001029057",
    "tab": "PCTBIBLIO"
}
```

```python
r = requests.post(url, data)
```

```python
r.status_code
```

```python
r.headers.get("content-type","unknown")
```

```python
response = requests.request(
        method="GET",
        url=url,
        params=data #(params should be a python object)
    )
soup = BeautifulSoup(response.text, 'html.parser')
soup
```
